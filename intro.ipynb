```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
import scipy
```

```python
x = np.arange(128) - 128/2
X,Y = np.meshgrid(x,x) 
f = (100*(np.hypot(X,Y)<32)).astype(np.uint8)
```

```python
plt.imshow(f)
```

```python
plt.gray()
plt.imshow(f)
```

```python
f.shape
```

```python
sift = cv2.SIFT_create()
kp = sift.detect(f,None)
 
img=cv2.drawKeypoints(f,kp,None)

plt.imshow(img)
```
