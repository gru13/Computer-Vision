```python
import cv2
```

Reading a image

```python
image = cv2.imread('./dataset/nachtwacht2.jpg')
```

Converting Color (Grayscale, BGR to RGB)

```python
# Convert to grayscale
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Convert from BGR to RGB
rgb_image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
```

Displaying an Image

```python
import matplotlib.pyplot as plt

plt.imshow(rgb_image)
plt.axis('off')  # Hide the axes
plt.show()
```

saving Image

```python
cv2.imwrite('grayscale_image.jpg', gray_image)
```

Flipping an Image

```python

flipped_image = cv2.flip(rgb_image, 1)
flipped_image_vertical = cv2.flip(rgb_image, 0)

plt.subplot(1, 3, 1)
plt.imshow(rgb_image)
plt.axis('off')
plt.title("Original Image")

plt.subplot(1, 3, 2)
plt.imshow(flipped_image)
plt.axis('off')
plt.title("Horizontally Flipped")

plt.subplot(1, 3, 3)
plt.imshow(flipped_image_vertical)
plt.axis('off')
plt.title("Vertically Flipped")

plt.show()
```

Cropping an Image

```python
cropped_image = rgb_image[100:300, 200:400]
plt.imshow(cropped_image)
plt.axis("off")
```

histogram equalization (for grayscale images)

```python
# Apply binary thresholding
_, thresholded_image = cv2.threshold(gray_image, 127, 255, cv2.THRESH_BINARY)

plt.imshow(thresholded_image)
```

Canny edge detection

```python
edges = cv2.Canny(image, 100, 200)

plt.imshow(edges)
```

Drawing Shapes on an Image

```python
# Draw a rectangle (start_point, end_point, color, thickness)
cv2.rectangle(image, (50, 50), (200, 200), (0, 255, 0), 2)

# Draw a circle (center, radius, color, thickness)
cv2.circle(image, (150, 150), 50, (255, 0, 0), -1)

plt.imshow(image)
```
