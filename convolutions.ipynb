```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
from scipy.signal import convolve2d
```

```python

# Path to your image
image_path = "./dataset/colortrui.png"  # Replace with your image path

# Load and preprocess the image
image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)  # Load grayscale image
image = cv2.resize(image, (128, 128))  # Resize the image
image = image / 255.0  # Normalize to [0, 1]
```

```python

# Define kernels
kernels = {
    "Blur": np.array([[1, 1, 1], [1, 1, 1], [1, 1, 1]]) / 9.0,
    "Edge Detection": np.array([[-1, -1, -1], [-1, 8, -1], [-1, -1, -1]]),
    "Horizontal Lines": np.array([[-1, -1, -1], [2, 2, 2], [-1, -1, -1]]),
    "Vertical Lines": np.array([[-1, 2, -1], [-1, 2, -1], [-1, 2, -1]]),
}
```

```python
# Apply convolution and visualize results
fig, axes = plt.subplots(1, len(kernels) + 1, figsize=(15, 5))
axes[0].imshow(image, cmap='gray')
axes[0].set_title("Original Image")
axes[0].axis('off')

for i, (name, kernel) in enumerate(kernels.items(), 1):
    convolved_image = convolve2d(image, kernel, mode='same', boundary='fill', fillvalue=0)
    axes[i].imshow(convolved_image, cmap='gray')
    axes[i].set_title(name)
    axes[i].axis('off')

plt.tight_layout()
plt.show()
```
